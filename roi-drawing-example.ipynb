```python
import numpy as np
import matplotlib.pyplot as plt
from fish.image.roi import RoiDrawing
%matplotlib notebook
```

```python
data = np.random.randn(100,100,100)
```

```python
fig, axs = plt.subplots()
sample = data.mean(0)
axs.imshow(sample)
drawing = RoiDrawing(axs, sample)
```

```python
fig, axs = plt.subplots(nrows=1,ncols=len(drawing.rois), figsize=(9,3))
for ind, r in enumerate(drawing.rois):
    axs[ind].imshow(r.get_mask())
```

```python
fig, axs = plt.subplots()
coords = np.where(drawing.rois[1].get_mask())
axs.plot(data[:,coords[0], coords[1]].mean(1))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Button
import roi
%matplotlib notebook
```

```python
fig, axs = plt.subplots(figsize=(8,4))
data = np.random.randn(100,100,100)
sample = data.mean(0)
axs.imshow(sample, cmap='gray')
drawing = roi.RoiDrawing(axs, sample)
ax_butt = plt.axes([0.7, 0.05, 0.1, 0.075])
wipe_butt = Button(ax_butt, 'Wipe')
wipe_butt.on_clicked(drawing.wipe)

ax_decr = plt.axes([.7, .15, .1, .075])
decr_button = Button(ax_decr, 'Prev ROI')
decr_button.on_clicked(drawing.focus_decr)

ax_incr = plt.axes([.7, .25, .1, .075])
incr_button = Button(ax_incr, 'Next ROI')
incr_button.on_clicked(drawing.focus_incr)
```

```python
fig, axs = plt.subplots(nrows=1,ncols=len(drawing.rois), figsize=(9,3))
for ind, r in enumerate(drawing.rois):
    axs[ind].imshow(r.get_mask())
```

```python
fig, axs = plt.subplots()
coords = np.where(drawing.rois[1].get_mask())
axs.plot(data[:,coords[0], coords[1]].mean(1))
plt.show()
```
